```python
import pandas as pd
df_dataset = pd.read_csv("eigen_train_files_with_gt.txt",header=None, sep=' ', names=['left', 'right', 'info'])
df_dataset.head()
```

```python
# use only 100 images taken by left camera
files_paths = df_dataset.left[:100]
files_paths
```

```python
#create local pathes
import os
if not os.path.exists("dataset"):
    os.makedirs("dataset")
else :
    print("folder exisit")
```

```python
# copy 100 images to create custom annotation
for path in files_paths:
    path = "../dataset/raw/" + path
    #print(path)
    command = "cp " + path + " ./dataset/"
    #print(command)
    os.system(command)
```
